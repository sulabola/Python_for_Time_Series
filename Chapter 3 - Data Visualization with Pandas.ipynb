Pandas can visualize data in a series or a dataframe. These are built directly off a library called 'Matplotlib'

Note: Here, more focus is given to time series analysis. However, 'Matplotlib' documentation has more options for customizing plots.

### Visualizing Data with Pandas

```python
import pandas as pd
import matplotlib.pyplot as plt  # use pyplot explicitly
%matplotlib inline
### The above code tells Jupyter that I want to see the plots within the notebook
### Note: This is only for Jupyter notebook users
```

```python
### Load data
df1 = pd.read_csv('Chapter_3_Pandas_Example1.csv', index_col=0)
df1.head()
```

```python
### Load data
df2 = pd.read_csv('Chapter_3_Pandas_Example2.csv')
df2.head()
```

#### Histogram

A histogram shows the underlying distribution of quantitative continuous data.

```python
df1['A'].plot.hist()
### Ensure it displays
plt.show()
### Each time we run the code, the plot color may change.
```

```python
### In most cases we may want to scale the outer box to fit the range.
### Also, we may want to see the lines separating each bar.

df1['A'].plot.hist(edgecolor='k') ### Black edges. Alternative: edgecolor='black'
plt.show()

### More colors available in matplotlib can be found in: https://matplotlib.org/stable/api/colors_api.html
```

```python
### To scale the outer box (Box starts at the end of the bars)
df1['A'].plot.hist(edgecolor='k').autoscale(enable=True,axis = 'both', tight=False)
plt.show()

### Note:
### enable=True and axis = 'both' - Handles where the outer box starts on the left and right sides
### tight=False - Handles where the outer box starts on the top
```

```python
### Change the number of bins

df1['A'].plot.hist(edgecolor='k', bins=20) ### Black edges
plt.show()
```

```python
### Introduce a grid to the plot

df1['A'].plot.hist(edgecolor='k', bins = 20, grid = True)
plt.show()
```

#### Barplot

A plot to represent categorical/discrete data.

```python
df2.head()
```

```python
df2.plot.bar()
plt.show()

### Note: Python automatically considers the index as a category. "e" is not included as it is a categorical variable.
```

```python
#### The barplot above has a bar for each category (column). We can also stack them.

df2.plot.bar(stacked=True)
plt.show()

### Note: We can also add arguments in the barplots to mention what is the x-axis variable and what is the y-axis variable.
```

```python
### We can also make a horizontal barplot

df2.plot.barh(stacked=True)
plt.show()
```

### Line Plot

```python
df2.plot.line(y='a')
plt.show()
```

```python
### Change figure size
df2.plot.line(y='a',figsize=(10,4)) ### This creates 10 inch * 4 inch graph
plt.show()
```

```python
### Change the thickness of the line
df2.plot.line(y='a', figsize=(10,4), lw=4)
plt.show()
```

```python
### We can also include multiple lines in the same plot
df2.plot.line(y=['a','b','c'], figsize=(10,4), lw=4)
plt.show()

### Note: If you do not mention 'y' values, it considers all the columns.
```

#### Area Chart

An area chart or area graph displays graphically quantitative data. It is based on the line chart. The area between axis and line are commonly emphasized with colors, textures and hatchings. Commonly, one compares two or more quantities with an area chart.

```python
df2.plot.area()
plt.show()
```

```python
### WE can also make this graph a little bit transparent.

df2.plot.area(alpha=0.4)
plt.show()
```

```python
### We can also remove stacking here

df2.plot.area(stacked=False, alpha=0.4)
plt.show()
```

#### Scatter Plot

A scatter plot can be used to compare two quantitative variables.

```python
df1.plot.scatter(x='A',y='B')
plt.show()
```

```python
### Add color to the scatter plot
### 'color by'
### We can change the color of points based on column 'C' or 'D'

df1.plot.scatter(x='A',y='B', c='C') ### This will add colors based on C
plt.show()
```

```python
df1.plot.scatter(x='A',y='B', c='C', cmap = 'coolwarm')
plt.show()
```

```python
### We can also change the size of the points based on other column values
### We can add the argument 's' to change the size. But it has to be a vector.

df1.plot.scatter(x='A',y='B', c='C', s = df1['C'])
plt.show()

### We get a warning here because we have negative values in column 'C'. Also, the values of 'C' are pretty small. Try: s = df1['C']*50
```

```python
### We can also add an alpha variable to make them see through

df1.plot.scatter(x='A',y='B', c='C', s = df1['C']*25, alpha=0.3)
plt.show()
```

#### Hexagonal Bin Plot

It is a 2D histogram plot, in which the bins are hexagons and the color represents the number of data points within each bin.

When we have a dataframe with a ton of data points, we may not be able to present them clearly. Also, the argument 'alpha' may not be helpful due to the large volume of data points. Then, hexagonal bin plots are helpful.

```python
### Let's create a dataframe first
import numpy as np
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
df.head()
```

```python
### Let's get a scatter plot
df.plot.scatter(x='a',y='b')
plt.show()

### In the following plot, it is hard to see how many points are stacked on top of each other.
```

```python
### Instead of points, we will draw a hexagon across, and depending on how many points are in each hexagon, it will be darker.

df.plot.hexbin(x='a',y='b', gridsize=25)
plt.show()

### without the argument 'gridsize', the size of the hexagon will be smaller.

### Note: There are more points near the centre (darker hexagons).
```

```python
### We can also include color mapping on this plot.

df.plot.hexbin(x='a',y='b', gridsize=25, cmap = 'coolwarm')
plt.show()
```

```python
df.plot.hexbin(x='a',y='b', gridsize=25, cmap = 'Oranges')
plt.show()
```

### Customizing Plots Created with Pandas

Note that we would not be able to customize every feature of a plot using pandas, as it is an abstraction of 'Matplotlib'. In order to customize plots further, we may have to go for 'Matplotlib'.

```python
###import pandas as pd
###import matplotlib.pyplot as plt  # use pyplot explicitly
###%matplotlib inline
###df2 = pd.read_csv('Chapter_3_Pandas_Example2.csv')
df2.head()
```

```python
### First, we get a line plot.
df2.plot.line()
plt.show()

### Note: Default plot type for Pandas is a line plot. So we can just type 'df2.plot()' to get a line plot.
```

```python
### Let's focus on one column, column 'C'.

df2['c'].plot.line()
plt.show()
```

```python
### Modify the figure size

df2['c'].plot.line(figsize=(20,3)) ### length = 20 inches, height = 3 inches
plt.show()
```

```python
### Modify the line type

df2['c'].plot.line(figsize=(10,3), ls='--')
plt.show()

### Default = '-'
### Other options: '-.' = dash and dotted line, ':' = dotted line
```

```python
### Change the color

df2['c'].plot.line(figsize=(10,3), ls='--', c='red')
plt.show()
```

```python
### Change line width

df2['c'].plot.line(figsize=(10,3), ls='--', c='red', lw=4)
plt.show()
```

#### Add a Title, X-axis Label, and Legend to a Plot

```python
title = "My Plot Title"
xlabel = "My X Data"
ylabel = "My Y Data"
```

```python
### Add axis labels

ax = df2['c'].plot.line(figsize=(10,3), ls='--', c='red', lw=4) ### ax for axis
ax.set(xlabel=xlabel, ylabel=ylabel) ### Alternatively we can directly mention names here
plt.show()
```

```python
### Add a title

ax = df2['c'].plot.line(figsize=(10,3), ls='--', c='red', lw=4, title = title)
ax.set(xlabel=xlabel, ylabel=ylabel)
plt.show()
```

```python
### Add a legend

df2.plot()
plt.show()

### Note: By default, it adds a legend. However, we can move it around using a location code.
```

```python
ax = df2.plot()
ax.legend(loc=0) ### with 0 it determines a location where it covers fewer data points.
plt.show()
```

```python
ax = df2.plot()
ax.legend(loc=1)
plt.show()

### Alternative:

#ax = df2.plot()
#ax.legend(loc='upper right')
#plt.show()
```

```python
### OR we can place the legend outside the plot.

ax = df2.plot()
ax.legend(loc=0, bbox_to_anchor=(1.0,1.0)) ### How much we want to move our legend from location '0' (We can set different starting positions)
plt.show()
```
